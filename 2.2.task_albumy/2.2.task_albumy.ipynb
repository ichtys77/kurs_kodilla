```python
import pandas as pd
```

Wczytaj bazę danych najlepiej sprzedających się albumów wszechczasów

```python
url = 'https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551'
table = pd.read_html(url)
df = table[0]
df_final = df.iloc[1:,1:]
```

1. Zamień nagłówki kolumn na polskie odpowiedniki

```python
df_final.columns = ['TYTUŁ','ARTYSTA','ROK','MAX POZ']
df_final
```

2. Ilu pojedynczych artystów znajduje się na liście

```python
unique_artist = df_final['ARTYSTA'].nunique()
unique_artist
```

3. Które zespoły pojawiają się najczęściej na liście (TOP5)

```python
top_bands = df_final['ARTYSTA'].value_counts().head(5)
top_bands
```

4. Zmień nagłówki kolumn, tak aby każdy z nich rozpoczynał się od wielkiej litery, a pozostałe były wprowadzone małymi literami

```python
df_final.columns = [col.capitalize() for col in df_final.columns]
df_final
```

5. Wyrzuć z tabeli kolumnę ‘Max poz’:

```python
df_final = df_final.drop(columns=['Max poz'])
df_final
```

6. W którym roku wyszło najwięcej albumów znajdujących się na liście

```python
best_year = df_final['Rok'].value_counts().idxmax()
best_year
```

7. Ile albumów wydanych między 1960 a 1990 rokiem włącznie znajduje się na liście

```python
df_final['Rok'] = pd.to_numeric(df_final['Rok'], errors='coerce')
```

```python
albums_1960_1990 = df_final[(df_final['Rok'] >= 1960) & (df_final['Rok'] <= 1990)].shape[0]
albums_1960_1990
```

8. W którym roku wydany został najmłodszy album na liście

```python
year_of_the_newest_album = df_final['Rok'].max()
year_of_the_newest_album
```

9. Przygotuj listę najwcześniej wydanych albumów każdego artysty, który znalazł się na liście:

```python
the_earliest_albums = df_final.loc[df_final.groupby('Artysta')['Rok'].idxmin()]
the_earliest_albums
```

```python
the_earliest_albums.to_csv('the_earliest_albums.csv', index=False)
```
