```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

1. Ile będzie wynosiła orientacyjna cena mieszkania za 5 lat?

```python
# założenia
actual_price = 120000  # aktualna cena mieszkania
rate = 0.05  # roczny wskaźnik wzrostu cen
years = 5  # liczba lat

# Obliczanie przyszłej wartości
future_price = actual_price * (1 + rate) ** years
print("Orientacyjna cena mieszkania za 5 lat:", future_price)
```

2. Ile musisz wpłacać do banku każdego miesiąca, aby przy przedstawionej ofercie uzbierać na mieszkanie w ciągu 5 lat?

```python
# założenia
annual_rate = 0.12  # Roczna stopa procentowa
monthly_rate = annual_rate / 12  # Miesięczna stopa procentowa
months = years * 12  # Liczba miesięcy

# Obliczanie miesięcznej wpłaty
monthly_payment = npf.pmt(monthly_rate, months, 0, -future_price)
print("Miesięczna wpłata:", monthly_payment)
```

3. wykres

```python
# Wzrost ceny mieszkania
house_prices = [actual_price * (1 + rate/12)**m for m in range(months+1)]
```

```python
# Wzrost wartości lokaty
deposit_values = [npf.fv(monthly_rate, m, -monthly_payment, 0) for m in range(months+1)]
```

```python
plt.figure(figsize=(8, 4))
plt.plot(house_prices, label='Cena mieszkania')
plt.plot(deposit_values, label='Wartość lokaty')
plt.xlabel('Miesiące')
plt.ylabel('Kwota (zł)')
plt.title('Wzrost ceny mieszkania i wartość lokaty')
plt.legend()
plt.grid(True)
plt.show()
```
